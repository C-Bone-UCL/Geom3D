## Example Usage of Fragment Scaffold Split

### Import dependancies and load dataset / Config File

```python
from geom3d.train_models import *
import torch
import copy

from geom3d.utils import fragment_scaffold_split
from geom3d.utils import database_utils


import os
os.chdir('/rds/general/user/cb1319/home/GEOM3D/Geom3D/')
config_dir = os.getcwd()+"/datasize_study/SchNet_opt_combined_80000"
config = read_config(config_dir)

dataset_path = "/rds/general/user/cb1319/home/GEOM3D/Geom3D/datasize_study/80000dataset_combined.pt"
dataset = torch.load(dataset_path)
```

### Example of Fragment Scaffold split for all 6 clusters

```python
import importlib
from geom3d.utils import fragment_scaffold_split
from geom3d.utils.fragment_scaffold_split import *
import pandas as pd

importlib.reload(fragment_scaffold_split)

# Initialize DataFrame
df_cluster_keys = pd.DataFrame(columns=[f"Cluster_{i}_keys" for i in range(1, 8)])

# Values to change the config["test_set_fragment_cluster"] to
values = [1, 2, 3, 4, 5, 6]

# Loop over each value in values
for value in values:
    # Change the value of config["test_set_fragment_cluster"] to the current value
    config["test_set_fragment_cluster"] = value
    
    # Run the fragment_scaffold_splitter function to get the cluster keys
    cluster_keys = fragment_scaffold_splitter(dataset, config)
    
    # Add cluster keys to the DataFrame
    df_cluster_keys[f"Cluster_{value}_keys"] = [",".join(cluster_keys)]

    print("Number of test samples for cluster", value, ":", len(cluster_keys))

```
